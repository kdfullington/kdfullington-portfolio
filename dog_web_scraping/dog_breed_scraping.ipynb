# Dog Breed Webscraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
# Request information from webpage
page = requests.get('https://www.dogbreedslist.info/all-dog-breeds')
breed_soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Note the error below
breed_soup
```

```python
# I needed to add headers so the website would think a person was accessing it - otherwise a "Not Acceptable!...This error was
# generated by Mod_Security"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',}
page = requests.get('https://www.dogbreedslist.info/all-dog-breeds', headers = headers)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# View the soup
soup
```

```python
# Create list of the div elements
# There should be one element for each dog in the list (expecting 20)
list = soup.findAll('div', attrs = {'class':'list'})
print(len(list))
```

```python
# View the list
print(list)
```

```python
# Go through the soup and select the 'div' elements, based on the class 'list_1', which based on above, contains the info we want
breeds = soup.findAll('div', attrs = {'class': 'list-1'}, limit = None)
data = [[x.text for x in y.findAll('p')] for y in breeds]
# Create a dataframe
df = pd.DataFrame(data)
```

```python
# Draw the information from the rest of the pages on the website (there are five pages of results)
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',}
page = requests.get('https://www.dogbreedslist.info/all-dog-breeds/page-2.html', headers = headers)
soup_2 = BeautifulSoup(page.content, 'html.parser')

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',}
page = requests.get('https://www.dogbreedslist.info/all-dog-breeds/page-3.html', headers = headers)
soup_3 = BeautifulSoup(page.content, 'html.parser')

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',}
page = requests.get('https://www.dogbreedslist.info/all-dog-breeds/page-4.html', headers = headers)
soup_4 = BeautifulSoup(page.content, 'html.parser')

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',}
page = requests.get('https://www.dogbreedslist.info/all-dog-breeds/page-5.html', headers = headers)
soup_5 = BeautifulSoup(page.content, 'html.parser')
```

```python
# Create lists from each of those pages and check the length (each should be 20)
list_2 = soup_2.findAll('div', attrs = {'class':'list'})
print(len(list))

list_3 = soup_3.findAll('div', attrs = {'class':'list'})
print(len(list))

list_4 = soup_4.findAll('div', attrs = {'class':'list'})
print(len(list))

list_5 = soup_5.findAll('div', attrs = {'class':'list'})
print(len(list))
```

```python
# Create dataframes from each of the lists
breeds_2 = soup_2.findAll('div', attrs = {'class': 'list-1'}, limit = None)
data_2 = [[x.text for x in y.findAll('p')] for y in breeds]

df_2 = pd.DataFrame(data_2)

breeds = soup_3.findAll('div', attrs = {'class': 'list-1'}, limit = None)
data_3 = [[x.text for x in y.findAll('p')] for y in breeds]

df_3 = pd.DataFrame(data_3)

breeds = soup_4.findAll('div', attrs = {'class': 'list-1'}, limit = None)
data_4 = [[x.text for x in y.findAll('p')] for y in breeds]

df_4 = pd.DataFrame(data_4)

breeds = soup_5.findAll('div', attrs = {'class': 'list-1'}, limit = None)
data_5 = [[x.text for x in y.findAll('p')] for y in breeds]

df_5 = pd.DataFrame(data_5)
```

```python
# Append the new dataframes to the end of the original dataframe
df = df.append(df_2)
df = df.append(df_3)
df = df.append(df_4)
df = df.append(df_5)
```

```python
df
```

```python
# Change the column names to be descriptive and easy to use in future code
df.rename({0: 'breed_name', 1: 'popularity_rank', 2:'hypoallergenic', 3:'intelligence_rank', 4:'price_range'}, axis = 1, inplace = True)
```

```python
df
```

```python
# Remove the '#' characters from the popularity_rank column
df['popularity_rank'] = df.popularity_rank.str.lstrip('#')
```

```python
df
```

```python
# Strip the word 'Prices:' from the price_range column
df['price_range'] = df.price_range.str.lstrip(' Prices: ')
```

```python
df
```

```python
# Split the price range column on '-' and create two new columns with each half
df[['min_price','max_price']] = df.price_range.str.split("-",expand=True,)
```

```python
df
```

```python
# Remove the '$' characters from the front of min_price and max_price
df['min_price'] = df.min_price.str.lstrip('$')
df['max_price'] = df.max_price.str.lstrip(' $')
```

```python
df
```

```python
# Change data type in min_price and max_price to integers
df['min_price'] = df['min_price'].astype(str).astype(int)
df['max_price'] = df['max_price'].astype(str).astype(int)
```

```python
# create excel writer object
writer = pd.ExcelWriter('breeds_df.xlsx')
# write dataframe to excel
df.to_excel(writer)
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
```
