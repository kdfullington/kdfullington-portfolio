## Dog Breed Data from Petfinder

The below is the documentation for finding specific types of animals.

Petfinder.animals([animal_id=None][, animal_type=None][, breed=None][, size=None][, gender=None][, age=None][, color=None][, coat=None][, status=None][,
name=None][, organization_id=None][, location=None][, distance=None][,
sort=None][, results_per_page=None][, pages=None][, return_df=False])


```python
# Create dataframe for ten pages worth of dog results
pf_dogs = pf.animals(animal_type = 'dog', status = ['adopted', 'adoptable'], pages = 10, return_df = True)
```

```python
# Set to show all columns and display the top rows of the pf_dogs dataframe
pd.set_option('display.max_columns', None)
pf_dogs.head()
```

```python
# Check the number of results
len(pf_dogs.index)
```

```python
pf_dogs.columns = pf_dogs.columns.str.replace(".", "_")
```

```python
# Display top rows of pf_dogs dataframe to verify that columns have been updated
pf_dogs.head()
```

```python
len(pd.isnull(pf_dogs))
```

```python
# Rows with null values
pf_dog_nulls = pd.isnull(pf_dogs)
len(pf_dog_nulls)
```

Remove rows that are all null values and check length of dataframe.

```python
pf_dogs = pf_dogs.dropna(how = 'all')
pf_dog_nulls = pd.isnull(pf_dogs)
len(pf_dog_nulls)
```

We can see that all 200 rows have at least one null value. I'm most interested in the breeds, so let's remove the rows that have null values for breeds_primary.

```python
pf_dogs = pf_dogs.dropna(subset = ['breeds_primary'])
```

```python
len(pf_dogs)
```

```python
pf_dogs.head()
```

Looks like all 200 rows have a valid value in breeds_primary, which is good.

```python
# Check for and list duplicates in the dataframe
pf_dogs[pf_dogs.id.duplicated(keep = False)]
```

We found one row that was duplicated. Let's drop that row.

```python
# Drop row by index value (obtained above) 180
pf_dogs = pf_dogs.drop(labels = 122, axis=0)
# Check for duplicates again and list them.
pf_dogs[pf_dogs.id.duplicated(keep = False)]
```

```python
# Check new length of dataframe
len(pf_dogs)
```

Now we see there are no duplicates in the dataset.

I want to create a new column with length of stay.

```python
# Create new dataframe with ONLY dogs whose status is 'adopted'
pf_dogs_adopted = pf.animals(animal_type = 'dog', status = 'adopted', before_date = '2022-01-01', after_date = '2021-01-01', pages = 15, return_df = True)
```

```python
# Check the top rows of the dataframe
pf_dogs_adopted.head()
```

```python
# pf_dogs_adopted['status_changed_at'] = pf_dogs_adopted['status_changed_at'].str.replace('[+000]', '')
# pf_dogs_adopted['published_at'] = pf_dogs_adopted['published_at'].str.replace('[+000]', '')
pf_dogs_adopted['status_changed_at'] = pf_dogs_adopted['status_changed_at'].map(lambda x: x.rstrip('+000'))
pf_dogs_adopted['published_at'] = pf_dogs_adopted['published_at'].map(lambda x: x.rstrip('+000'))

pf_dogs_adopted['status_changed_at'] = pf_dogs_adopted['status_changed_at'].str.split('T').str[0]
pf_dogs_adopted['published_at'] = pf_dogs_adopted['published_at'].str.split('T').str[0]
```

```python
pf_dogs_adopted.head()
```

```python
pf_dogs_adopted['status_changed_at'] = pd.to_datetime(pf_dogs_adopted['status_changed_at'], format = '%Y-%m-%d')
pf_dogs_adopted['published_at'] = pd.to_datetime(pf_dogs_adopted['published_at'], format = '%Y-%m-%d')
```

```python
pf_dogs_adopted.head()
```

```python
pf_dogs_adopted['length_of_stay'] = pf_dogs_adopted['status_changed_at'].sub(pf_dogs_adopted['published_at'], axis=0)
pf_dogs_adopted.head() 
```

```python
# create excel writer object
writer = pd.ExcelWriter('petfinder_df.xlsx')
# write dataframe to excel
pf_dogs_adopted.to_excel(writer)
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
```
