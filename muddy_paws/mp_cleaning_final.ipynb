```python
# Function to get sentiment score
def get_sentiment_score(text):
    if text is not None:
        blob = TextBlob(text)
        return blob.sentiment.polarity
    else:
        return None
    
# Apply the sentiment score function to create a new column
df['sentiment_score'] = df.temperament.apply(get_sentiment_score)
```
