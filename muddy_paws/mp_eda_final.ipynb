```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.length_of_stay_days.max()
```

```python
# Set up the figure size
plt.figure(figsize=(15, 6))

# Set up the histogram
plt.hist(df.length_of_stay_days, bins=np.arange(min(df.length_of_stay_days), max(df.length_of_stay_days) + 1) - 0.5, edgecolor='blue')

# Add labels and title
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Frequency')
plt.title('Histogram of Length of Stay')


# Customize grid and background colors
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().set_facecolor('#f3f3f3')


# Display the histogram
plt.show()
```

```python
# Count the occurrences of 'yes' and 'no' in the 'bully_breed' column
counts = df['bully_breed'].value_counts()

# Create a bar chart
plt.bar(counts.index, counts.values, color=['red', 'black'])

# Add labels above each bar
for i, count in enumerate(counts.values):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

    
plt.xlabel('Bully Breed')
plt.ylabel('Count')
plt.title('Distribution of Bully Breed (yes/no)')
plt.show()
```
