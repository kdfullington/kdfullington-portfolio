```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('book_w_images.csv')
```

```python
df.head(2)
```

```python
df = df.drop(['genre_0', 'genre_1', 'genre_2', 'genre_3', 'genre_4', 'genre_5', 'genre_6', 'genre_7', 'genre_8', 'genre_9', 'genre_10', 'genre_11', 
              'genre_12', 'genre_13', 'genre_14', 'genre_15', 'genre_16', 'genre_17', 'genre_18', 'genre_19'], axis = 1)
```

```python
df.head(2)
```

```python
# Set the size of the figure
plt.figure(figsize = (12, 6))

# Plotting the histogram using Pandas and Matplotlib
df.pages.plot(kind = 'hist', bins = range(min(df.pages), max(df.pages) + 1), edgecolor = 'blue')

# Adding labels and title
plt.xlabel('Number of Pages')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Pages')

# Display the plot
plt.show()
```

```python
df.pages.describe()
```

```python
df = df[df.pages > 0]
```

```python
# Set the size of the figure
plt.figure(figsize = (12, 6))

# Plotting the histogram using Pandas and Matplotlib
df.pages.plot(kind = 'hist', bins = range(min(df.pages), max(df.pages) + 1), edgecolor = 'blue')

# Adding labels and title
plt.xlabel('Number of Pages')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Pages')

# Display the plot
plt.show()
```

```python
# Set the size of the figure
plt.figure(figsize = (12, 6))

# Plotting the histogram using Pandas and Matplotlib
df.rating.plot(kind = 'hist', bins = 20, edgecolor = 'blue')

# Adding labels and title
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Histogram of Ratings')

# Display the plot
plt.show()
```

```python
# Set the size of the figure
plt.figure(figsize = (12, 6))

# Plotting the histogram using Pandas and Matplotlib
df.reviews.plot(kind = 'hist', bins = 200, edgecolor = 'blue')

# Adding labels and title
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Reviews')

# Display the plot
plt.show()
```
